### Data handling

- 강남구 , 강남을 통일해 준다. 
- 명시적으로 data을 대입해 주어야 원 데이터가 수정된다.


```python
data["지점"]= data["지점"].replace("강남구", "강남")
data
```

```python
# NaN(Not a Number), null  / 빈값 임의 값으로 채우기 
data["몸무게"] =  data["몸무게"].fillna(75.8)

data
```

```python
# 시계열 데이터에 요긴하게 쓰인다. 

data["몸무게(prev)"] = data["몸무게"].shift(+1)
data["몸무게(next)"] = data["몸무게"].shift(-1)

data
```

```python
# pandas 에게  년, 월, 일 등 날짜 데이터로 알려주기  
# 판다스는 날짜를 인식하지 못한다. 
# 명시적으로 일러줘야 한다. ㅠㅠㅠ
data["측정일"] = pd.to_datetime(data["측정일"])
data["측정일"].dt.year
```
