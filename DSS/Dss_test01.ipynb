### 1. 할인 가격
기름값을 계산하라. 많이 사면 할인을 해준다. 

- 처음 20갤론까지: 갤론 당 90센트 
- 그 다음 40갤론은: 갤론 당 80센트 
- 그외 나머지 : 갤론 당 75센트

예를 들어 35갤론을 구입했다면, 가격은 30달러 (20*90 + 15* 80)

```python
# 나머지 75 갤런을 계산하는 법을 추가해주세요. 
gallon = 80
if gallon <= 20:
    price = 90 * gallon
elif gallon <= 60:
    price = 90 * 20 + 80 * (gallon - 20)
else:    
    price = 90 * 20 + 80 * 40 + 75 *(gallon - 60)
print(gallon, price)    
```

369게임은 다음과 같습니다. 
1. 1 부터 시작해서 숫자를 하나씩 센다. 
2. 숫자에 3이나, 6이나 9가 있을경우 박수(clap)을 친다. 
3. 지정한 최대 숫자까지 도달하면 프로그램은 종료한다. 

```python
# 박수를 여러번 치는 경우를 고려해서 계산 
# 33은 박수를 두 번 쳐야 한다. 
# 문자열의 count method를 사용해 보세요. 

count = 334
for n in range(count):
    n_text = str(n)
    n_clap = n_text.count("3") + n_text.count("6") + n_text.count("9")
    
    if "3"  in n_text or "6"  in n_text or "9"  in n_text :
        print(n, "clap!" * n_clap)
    else:
        print(n)
        
    
#     if (n == 3) or (n == 6) or (n == 9):
#         print(i,"clap!")
#     else:
#         print(i)
```
