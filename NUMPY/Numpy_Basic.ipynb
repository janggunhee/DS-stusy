## 배열의 인덱싱과 슬라이싱 

### 1차원 배열의 인덱싱과 슬라이싱 : list와 동일한 방법 

```python
d = np.arange(12)
print(d)
print(d[0], d[1], d[2], d[-1])          # 배열의 인덱싱 

print(d[2:8], d[:-1], d[::3], d[::-1])  # 배열의 슬라이싱

d[0] = 10                               # list처럼 요소 수정이 가능
print(d)

```

### 2차원 배열의 인덱싱과 슬라이싱 

```python
d = np.arange(12).reshape(3,4)
print(d, d.shape)

# 인덱싱 
print(d[0,0])
print(d[0],[0])
print(d[1],[1])
print(d[2],[2])
print(d[2,3])


#슬라이싱 
print(d[0])
print("")
print(d[0:3])
print("")
print(d[2:3])
print("")
print(d[:, :-1])  # 마지막 열을 제외 
print("")
print(d[::2, ::2])
print("")
d[::2, ::3]
print(d)
print("")
d[::2, ::3] = 30  # 한번에 슬라이싱된 요소 전체를 변경할 수 있다.
print(d)
print("")
d[:, :] = -1
print(d)
```

### 3 차원 배열의 인덱싱과 슬라이싱 
### data[면, 행, 열]


```python
d = np.arange(12).reshape(2,2,3)
print(d)


# 인덱싱 
print(d[1, 0 , 0])
print(d[1][0][1])
print(d[0][1][2])



# 슬라이싱 
print(d[:, :-1, :-1])   # (2, 1, 2)
print(d[:, 1:, 1:])     # ( 모든 , 0번행 스킵, 0번 열 스킵)
print(d[:, -1, -1])     # d(  , 마지막 행, 마지막 열)  (2,)
print(d[:, :, -1])      # (2,2)  
```

## boolean indexing : 불린 인덱싱 : 조건식을 사용한 인덱싱, 조건 검색 필터를 사용한 추출 

```python
a = np.array([1,2,3,4,5,6])
print(a)
print(a>2)        # [False False  True  True  True  True]
print(a[a>2])

# r = []
# for k in a:
#     if k > 2:
#       r.append(k)  
# print('result:',r)    

# True인 요소민 반환한다. 
boolean_indexing = np.array([True, False, True, False, True, False])
print(a[boolean_indexing])

print(a[a!=3])
print(a[(a>2) & (a <6)])
print(a[(a%2) == 0])       # 2의 배수만 필터링 
print(a[a < a.mean()])     # 평균보다 작은 값만 필터링 

# print(a[True])  # 모든 요소가 출력
# print(a[False]) # 빈 배열이 출력
```

### 배열의 형상 바꾸기 : reshape( ), 편평화 flatten( ), 전치 transpose( )

```python
# flatten() : 다차원 배열을 1차원으로 만든다
d = np.arange(12).reshape(3,4)
print(d)

f = d.flatten()    # (3,4)  --> (12,)
print(f,f.shape)

r = f.reshape(3,4)
print(r)


# 데이터 순서를 거꾸로 변경
print(d[::-1])       # 행의 순서를 거꾸로 변경
print(d[:,::-1])     # 열의 순서를 거꾸로 변경
print(d[::-1,::-1])  # 행과열의 순서를 거꾸로 변경


# transpose() : 전치 행렬, 행과 열을 서로 맞바꾸는 함수
t = d.transpose()
print(t,t.shape)    # (4,3)

# 임시로 1차원으로 변경하여 데이터를 채우고 다시 원래의 shape을 갖는다
d.flat = 0
print(d,d.shape)
```

```python
# racoon (오리너구리 사진 처리)
# view : 원본객체와 다른 별도의 메로리를 사용하는 사본 
# Copy : 원본 객체와 다른 별도의 메모리를 사용하는 사본 
```

```python
import scipy.misc
import matplotlib.pyplot as plt

face = scipy.misc.face()   # read_only 만 가능 
face2 = face.copy()        # 별도의 메모리에 복사본이 생성 , read & write가 가능 

f_view = face2.view()      # face2 원본 객체와 메모리를 공유하는 사본 
f_copy = face2.copy()      # face2 원본객체와 다른 별도의 메모리를 사용하는 사본 

# face2.flat = 0            #  하나의 색으로 반환
# f_view.flat = 0 
f_copy.flat = 0 

print(type(face))
print(face.shape)         # (768, 1024, 3)  , (y, x, rgb)
print(face.nbytes/1024)   # 2.3 Mbyte

plt.subplot(2,2,1)
plt.imshow(face2)

plt.subplot(2,2,2)
plt.imshow(f_view)

plt.subplot(2,2,3)
plt.imshow(f_copy)

plt.show()
```

### 배열 합치기 
- 수직(v, 행)으로 합치기 : vstack(), row_stack(), concatenate(..., axis=0)
- tnvud(h, 열)으로 합치기 : hstack(), column_stack(), concatenage(...,axis=1)


```python
# 수직으로 합치기 
a = np.arange(9).reshape(3,3)
b = a * 2
print(a)
print(b)

r = np.vstack((a,b))         # 수평
print(r, r.shape)   #(6,3)

c = np.row_stack((a,b))      # 열
print(c)

x = np.concatenate((a,b), axis = 0)    # 행, axis = 0
print(x )
```

### 수평으로 합치기 

```python
a = np.arange(9).reshape(3,3)
b = a * 2
print(a)
print(b)

r = np.hstack((a,b))            # 수평
print(r, r.shape)               #(6,3)

c = np.column_stack((a,b))      # 열
print(c)

x = np.concatenate((a,b), axis = 1)    # 열, axis = 1
print(x)
```

### 쪼개기 
- vsplit(배열, 행의 갯수) : 수직으로 쪼갠다. (행) , split(배열, 행의 갯수 , axis = 0) 
- hsplit(배열, 행의 갯수) : 수평으로 쪼갠다. (열) , split(배열, 열의 갯수 , axis=1)

```python
a = np.arange(12).reshape(4, 3)
print(a)

print(np.vsplit(a, 4))         # array 4개를  4개로 쪼갬 
print(np.split(a, 4, axis = 0))
```

```python
a = np.arange(12).reshape(3, 4)
print(a)

print(np.hsplit(a, 4))
print(np.split(a, 4, axis = 1))
```
