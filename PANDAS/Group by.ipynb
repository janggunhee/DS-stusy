## Group by

```python
import pandas as pd 
import numpy as np
```

```python
from numpy.random import seed
from numpy.random import rand
from numpy.random import randint

seed(42)

df = pd.DataFrame({ 'Weather' : ['cold', 'hot', 'cold', 'hot', 'cold', 'hot', 'cold'],
                    'Food' : ['soup', 'soup', 'icecream', 'chocolate', 'icecream', 'icecream', 'soup'],
                    'Price' : 10 * rand(7), 
                    'Number' : randint(1, 9, 7)})

df
```

```python
weather_group = df.groupby('Weather')
print(weather_group)


print(weather_group.first())   # cold 의 첫번째 행 ,hot의 첫번째 행
print(weather_group.last())    # cold 의 마지막 행 ,hot의 마지막 행
print('-'*40)
print(weather_group.mean())     # cold 와 hot 각각의 평균
print('-'*40)
print(weather_group.count())     # cold 와 hot 각각의 총 갯수
print('-'*40)

print(weather_group.describe()) # cold 와 hot 각각의 통계 요약
print('-'*40)

# i = 0
for name, group in weather_group:
#    i = i + 1
    print(i, name, '\n', group)
    print(type(group))
```

```python
wf_group = df.groupby(['Weather', 'Food'])
print(wf_group)
type(wf_group.groups)


wf_group.agg([np.mean, np.sum, np.min, np.max])
```

### agg : 에그리게이션 함수 
-  https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy

```python
gr_df = wf_group.agg([np.mean, np.sum, np.min , np.max])
type(gr_df)
gr_df.index # MultiIndex

gr_df['Price']['mean']
```

```python
df1 = pd.DataFrame(gr_df['Price']['mean'])
df2 = pd.DataFrame(gr_df['Number']['mean'])
df3 = pd.concat([df1,df2],axis = 1)
df3
```

```python
pd.DataFrame([gr_df['Price']['mean'], gr_df['Number']['sum']])
```
